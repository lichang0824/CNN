```python
import pandas as pd
comparison = pd.read_csv('comparison.csv')
```

```python
comparison
```

```python
dA = {3:'1', 5:'2'}
dB = {'ReLU':'1', 'Sigmoid':'2'}
dC = {1e-3:'1', 1e-4:'2', 1e-5:'3'}
dD = {5:'1', 10:'2', 15:'3', 20:'4', 25:'5', 30:'6'}
```

```python
comparison['A'] = comparison['kernel_size'].map(dA)
comparison['B'] = comparison['activation_fn'].map(dB)
comparison['C'] = comparison['learning_rate'].map(dC)
comparison['D'] = comparison['epochs_choice'].map(dD)
```

```python
comparison['combo'] = comparison['A'] + comparison['B'] + comparison['C'] + comparison['D']
```

```python
comparison
```

```python
oa = pd.read_csv('MOA_L36.csv')
```

```python
oa['combo'] = oa['combo'].astype(str)
```

```python
comparison['combo'][0]
```

```python
merged = pd.merge(left = comparison, right = oa, on = 'combo', how = 'right')
```

```python
merged.to_csv('merged.csv', index = False)
```
